### Bivariate: Two Variable Analysis

```python
plt.figure(figsize=(6, 3))
top_countries = df.groupby('State')['Profit'].sum().nlargest(5)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='flare', edgecolor='black')
plt.title(' Top 5 State with highest profits', fontsize=15)
plt.xlabel('State', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

## Conclusion: 
- The state of "California" has the highest profit.

```python
plt.figure(figsize=(6, 3))
top_countries = df.groupby('City')['Profit'].sum().nlargest(5)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='flare', edgecolor='black')
plt.title(' Top 5 cities with highest profits', fontsize=15)
plt.xlabel('City', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

## Conclusion: 
- "New York City" has the highest profit.

```python
plt.figure(figsize=(6, 3))
top_countries = df.groupby('Sub-Category')['Profit'].sum().nlargest(5)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='flare', edgecolor='black')
plt.title(' Top 5 sub-categories with their profits', fontsize=15)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

## Conclusion: 
- The "Copiers" has the highest profit.

```python
plt.figure(figsize=(6, 3))
top_countries = df.groupby('Ship Mode')['Profit'].sum().nlargest(5)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='flare', edgecolor='black')
plt.title(' Top 5 shipping mode with profits', fontsize=15)
plt.xlabel('Ship Mode', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

## Conclusion:
- "Standard Class" shipping has highest profit.
